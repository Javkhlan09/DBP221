```python
import pandas as pd 
import numpy as np
```

```python
xl = pd.ExcelFile('Бие даалт - 1.xlsx')
df = xl.parse('Ирц')
names = xl.parse('Ирц')['Оюутны нэр']
```

```python
df = xl.parse('Ирц')
a = df.T.apply(pd.value_counts)[:2].T
d = pd.concat([names,a],axis=1)
ipts = pd.Series(d['и']/32*10, name='Ирцийн оноо')
ipts = pd.concat([names, ipts], axis=1)
ipts
```

```python
df = xl.parse('Семинар')
a = df.T[4:25].apply(pd.value_counts).T/19*25
a = a.squeeze()
seminap = a.rename('Семинарын оноо')
seminap = pd.concat([names,seminap],axis=1)
seminap
```

```python
df = xl.parse('Шалгалт')
shalgalt = df.T[4:7].sum().rename('Шалгалтын оноо')
shalgalt = pd.concat([names, shalgalt], axis=1)
shalgalt

```

```python
df = xl.parse('Бие даалт')
biydaalt = df.T[4:6].sum().rename('Бие даалтын оноо')
biydaalt = pd.concat([names, biydaalt], axis=1)
biydaalt

```

```python
df = xl.parse('Хүмүүжил хандлага')
humuujil = df.T[4:8].sum().rename('Хүмүүжил хандлагын оноо')
humuujil = pd.concat([names, humuujil], axis=1)
humuujil

```

```python
df = xl.parse('Нэмэлт оноо')
nemelt = df.T[4:5].T.squeeze()
nemelt = pd.concat([names, nemelt], axis=1)
nemelt

```
