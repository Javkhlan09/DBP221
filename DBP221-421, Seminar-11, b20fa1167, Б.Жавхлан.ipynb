```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["Python","Java","PHP","JavaScript","C#","C++"])
y = np.array([12.90,11.80,1.50,2.30,6.40,7.73])
plt.bar(x, y, color = ['black', 'red', 'green', 'blue', 'yellow','cyan'])
plt.show()
```

```python
y = np.array([30.21,17.82, 5.84, 9.16,7.53,6.82 ])
mylabels = ["Python","Java","PHP","JavaScript","C#","C++"]

plt.pie(y, labels = mylabels)
plt.show() 
```
